```python
import yfinance as yf
import pandas as pd

# Extracting Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Saving to a CSV file (optional)
tesla_data.to_csv('tesla_stock_data.csv')

# Display the first few rows of the data
tesla_data.head()

import requests
from bs4 import BeautifulSoup

# Define the URL for Tesla's revenue data
url = "URL_CONTAINING_TESLA_REVENUE_DATA"

# Make a request to the website
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Parse the revenue data
revenue_table = soup.find_all('table')[0]  # Adjust based on the specific structure
tesla_revenue = pd.read_html(str(revenue_table))[0]

# Save to a CSV file (optional)
tesla_revenue.to_csv('tesla_revenue_data.csv')

# Display the first few rows of the data
tesla_revenue.head()


# Define the URL for GameStop's revenue data
url = "URL_CONTAINING_GAMESTOP_REVENUE_DATA"

# Make a request to the website
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Parse the revenue data
revenue_table = soup.find_all('table')[0]  # Adjust based on the specific structure
gamestop_revenue = pd.read_html(str(revenue_table))[0]

# Save to a CSV file (optional)
gamestop_revenue.to_csv('gamestop_revenue_data.csv')

# Display the first few rows of the data
gamestop_revenue.head()


import matplotlib.pyplot as plt

# Plotting Tesla stock data
plt.figure(figsize=(14, 7))
plt.plot(tesla_data.index, tesla_data['Close'], label='Tesla Stock Price')
plt.title('Tesla Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Stock Price ($)')
plt.legend()
plt.show()

# Plotting Tesla revenue data
plt.figure(figsize=(14, 7))
plt.plot(pd.to_datetime(tesla_revenue['Date']), tesla_revenue['Revenue'], label='Tesla Revenue')
plt.title('Tesla Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue ($)')
plt.legend()
plt.show()


# Plotting GameStop stock data
plt.figure(figsize=(14, 7))
plt.plot(gamestop_data.index, gamestop_data['Close'], label='GameStop Stock Price')
plt.title('GameStop Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Stock Price ($)')
plt.legend()
plt.show()

# Plotting GameStop revenue data
plt.figure(figsize=(14, 7))
plt.plot(pd.to_datetime(gamestop_revenue['Date']), gamestop_revenue['Revenue'], label='GameStop Revenue')
plt.title('GameStop Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue ($)')
plt.legend()
plt.show()
```
